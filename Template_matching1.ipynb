```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np


# Loading original image and templates
img = cv.imread(r"C:\Users\Lenovo\Desktop\airport.jpg")
#In cv.imread(r"C:\Users\Lenovo\Desktop\temp5.jpg", 0) 0 makes the photo gray
img_gray = cv.imread(r"C:\Users\Lenovo\Desktop\airport.jpg", 0)
temp1 = cv.imread(r"C:\Users\Lenovo\Desktop\temp1.jpg", 0)
temp2 = cv.imread(r"C:\Users\Lenovo\Desktop\temp2.jpg", 0)
temp3 = cv.imread(r"C:\Users\Lenovo\Desktop\temp3.jpg", 0)
temp4 = cv.imread(r"C:\Users\Lenovo\Desktop\temp4.jpg", 0)
temp5 = cv.imread(r"C:\Users\Lenovo\Desktop\temp5.jpg", 0)



#Using cv.imagetemplate to find matched templates and drawing founded templates in original image
w1, h1 = temp1.shape[::-1]
w2, h2 = temp2.shape[::-1]
w3, h3 = temp3.shape[::-1]
w4, h4 = temp4.shape[::-1]
w5, h5 = temp5.shape[::-1]
res1 = cv.matchTemplate(img_gray, temp1, cv.TM_CCOEFF_NORMED)
res2 = cv.matchTemplate(img_gray, temp2, cv.TM_CCOEFF_NORMED)
res3 = cv.matchTemplate(img_gray, temp3, cv.TM_CCOEFF_NORMED)
res4 = cv.matchTemplate(img_gray, temp4, cv.TM_CCOEFF_NORMED)
res5 = cv.matchTemplate(img_gray, temp5, cv.TM_CCOEFF_NORMED)
loc1 = np.where( res1 >= 0.5)
for pt1 in zip(*loc1[::-1]):
    cv.rectangle(img, pt1, (pt1[0] + w1, pt1[1] + h1), (0,0,255), 2)

loc2 = np.where( res2 >= 0.46)
for pt2 in zip(*loc2[::-1]):
    cv.rectangle(img, pt2, (pt2[0] + w2, pt2[1] + h2), (0,0,255), 2)

loc3 = np.where( res3 >= 0.5)
for pt3 in zip(*loc3[::-1]):
    cv.rectangle(img, pt3, (pt3[0] + w3, pt3[1] + h3), (0,0,255), 2)

loc4 = np.where( res4 >= 0.5)
for pt4 in zip(*loc4[::-1]):
    cv.rectangle(img, pt4, (pt4[0] + w4, pt4[1] + h4), (0,0,255), 2)

loc5 = np.where( res5 >= 0.5)
for pt5 in zip(*loc5[::-1]):
    cv.rectangle(img, pt5, (pt5[0] + w5, pt5[1] + h5), (0,0,255), 2)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Result image")
plt.show()

#We used 5 templates for airplaines(based on their direction)
```
