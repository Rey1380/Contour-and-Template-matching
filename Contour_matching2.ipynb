```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

#reading and showing the original image
img_rgb1 = cv.imread(r"C:\Users\Lenovo\Desktop\store.png")
plt.imshow(cv.cvtColor(img_rgb1, cv.COLOR_BGR2RGB))
plt.title("original image")
plt.show()
x=y=0
img_rgb2 = cv.imread(r"C:\Users\Lenovo\Desktop\store.png")
assert img_rgb1 is not None, "file could not be read, check with os.path.exists()"

# making the original image gray
img_gray = cv.cvtColor(img_rgb1, cv.COLOR_BGR2GRAY)


#turned gray image of template images
template1 = cv.imread(r"C:\Users\Lenovo\Desktop\temp1.png", cv.IMREAD_GRAYSCALE)
plt.imshow(cv.cvtColor(template1, cv.COLOR_BGR2RGB))
plt.title("Template1")
plt.show()
assert template1 is not None, "file could not be read, check with os.path.exists()"
template2 = cv.imread(r"C:\Users\Lenovo\Desktop\temp0.png", cv.IMREAD_GRAYSCALE)
plt.imshow(cv.cvtColor(template2, cv.COLOR_BGR2RGB))
plt.title("Template2")
plt.show()
assert template2 is not None, "file could not be read, check with os.path.exists()"

#finding templates in original image by using matchTemplate
w1, h1 = template1.shape[::-1]
w2, h2 = template2.shape[::-1]
res1 = cv.matchTemplate(img_gray, template1, cv.TM_CCOEFF_NORMED)
res2 = cv.matchTemplate(img_gray, template2, cv.TM_CCOEFF_NORMED)

#drawing founded templates in original image
threshold = 0.8
loc1 = np.where( res1 >= threshold)
for pt1 in zip(*loc1[::-1]):
    cv.rectangle(img_rgb1, pt1, (pt1[0] + w1, pt1[1] + h1), (0,0,255), 2)
    x=x+1

cv.imwrite('res1.png',img_rgb1)
plt.imshow(cv.cvtColor(img_rgb1, cv.COLOR_BGR2RGB))
plt.title("Founded1")
plt.show()
loc2 = np.where( res2 >= threshold)
for pt2 in zip(*loc2[::-1]):
    cv.rectangle(img_rgb2, pt2, (pt2[0] + w2, pt2[1] + h2), (0,0,255), 2)
    y=y+1
cv.imwrite('res2.png',img_rgb2)
plt.imshow(cv.cvtColor(img_rgb2, cv.COLOR_BGR2RGB))
plt.title("Founded2")
plt.show()
print("first template numbers in original image:",x)
print("second template numbers in original image:",y)



```
